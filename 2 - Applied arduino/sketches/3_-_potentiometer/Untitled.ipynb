```python
import serial

ser = serial.Serial("/dev/cu.usbserial-1430", baudrate=9600, timeout=0.1) # timeout in seconds
```

```python
import numpy as np
```

```python
int(ser.readline().decode('utf-8')[:-2])
```

```python
while ser.in_waiting>100:
    ser.read(ser.in_waiting)

data = []
while True:
    data.append(int(ser.readline().decode('utf-8')[:-2]))
```
